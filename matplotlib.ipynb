```python
# ==========================================
# STEP 1: MATPLOTLIB LIBRARY KO IMPORT KARNA
# ==========================================
# 'pyplot' matplotlib ka wo hissa hai jo graphs banata hai.
# Hum ise 'plt' ke naam se use karenge.
import matplotlib.pyplot as plt
import random # Random numbers banane ke liye

# ==========================================
# QUESTION 1: LINE PLOT (BASIC GRAPH)
# ==========================================

# Data points tayyar karna
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

# a. Data ko plot karna (Line banane ke liye)
# marker='o' se points par dots dikhenge
plt.plot(x, y, marker='o', color='blue', linestyle='--')

# b. Customization (Title aur Labels lagana)
plt.title("Simple Line Plot") # Graph ka heading
plt.xlabel("X-axis (Steps)") # Neeche wali line ka naam
plt.ylabel("Y-axis (Values)") # Khadi wali line ka naam
plt.grid(True) # Peeche box wali lines (grid) dikhane ke liye
plt.show() # Graph ko screen par dikhana


# ==========================================
# QUESTION 2: BAR GRAPH (STUDENTS MARKS)
# ==========================================

students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

# a. Bar graph banana
# plt.bar use hota hai khade dande (bars) banane ke liye
plt.bar(students, marks, color=['skyblue', 'orange', 'lightgreen', 'pink'])

# b. Customization
plt.title("Students Marks in a Subject")
plt.xlabel("Student Names")
plt.ylabel("Marks Scored")
plt.show()


# ==========================================
# QUESTION 3: PIE CHART (REVENUE DISTRIBUTION)
# ==========================================

regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# b. Highlight karna (Explode): 
# Sabse zyada revenue North America (45) ka hai, use thoda bahar nikalenge.
# (0.1, 0, 0, 0) ka matlab pehla hissa thoda cut-out dikhega.
explode = (0.1, 0, 0, 0) 

# a. Pie chart banana
# autopct='%1.1f%%' se percentage graph ke andar dikhegi
plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)

plt.title("Company Revenue by Region")
plt.show()


# ==========================================
# QUESTION 4: HISTOGRAM (FREQUENCY DISTRIBUTION)
# ==========================================

# 1 se 100 ke beech 1000 random numbers generate karna
# Ye list comprehension hai, simple loop jaisa
random_data = [random.randint(1, 100) for _ in range(1000)]

# Histogram banana
# 'bins=20' ka matlab hai data ko 20 groups mein baantna
plt.hist(random_data, bins=20, color='purple', edgecolor='black')

plt.title("Frequency Distribution of Random Numbers")
plt.xlabel("Value Range (1-100)")
plt.ylabel("How many times (Frequency)")
plt.show()
```
