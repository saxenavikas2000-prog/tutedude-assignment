```python
# ==========================================
# STEP 1: PANDAS LIBRARY KO IMPORT KARNA
# ==========================================
# 'import pandas as pd' ka matlab hai hum Pandas library use karenge 
# aur use short mein 'pd' bulayenge.
import pandas as pd

# ==========================================
# QUESTION 1: DATAFRAME BANANA AUR BASIC INFO
# ==========================================

# Ek dictionary banate hain jisme data hai
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

# pd.DataFrame se hum is data ko ek Table (DataFrame) mein badal denge
df = pd.DataFrame(data)

# a. Pehli 5 rows print karna
# df.head() function pehli 5 rows dikhata hai
print("1a. First 5 rows of the DataFrame:")
print(df.head())

# b. Age aur Salary ka summary statistics (Mean, Max, Min etc.) nikalna
# describe() function se humein count, mean, std deviation sab mil jata hai
print("\n1b. Summary statistics for Age and Salary:")
print(df[['Age', 'Salary']].describe())

# c. HR department ke employees ki average salary nikalna
# Pehle hum sirf 'HR' waale logo ko filter karenge
hr_employees = df[df['Department'] == 'HR']
# Phir unki salary ka mean (average) nikalenge
avg_hr_salary = hr_employees['Salary'].mean()
print(f"\n1c. Average salary of employees in HR: {avg_hr_salary}")


# ==========================================
# QUESTION 2: NAYI COLUMN 'BONUS' JODNA
# ==========================================

# Bonus column banane ke liye hum Salary ko 0.10 (yani 10%) se multiply karenge
df['Bonus'] = df['Salary'] * 0.10
print("\n2. DataFrame after adding 'Bonus' column:")
print(df)


# ==========================================
# QUESTION 3: FILTERING (AGE BETWEEN 25 AND 30)
# ==========================================

# Hum un logo ko chunenge jinki Age >= 25 hai AND Age <= 30 hai
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print("\n3. Employees aged between 25 and 30:")
print(filtered_df)


# ==========================================
# QUESTION 4: GROUPING BY DEPARTMENT
# ==========================================

# groupby() se hum data ko department wise alag karenge 
# aur ['Salary'].mean() se har group ki average salary nikalenge
dept_avg_salary = df.groupby('Department')['Salary'].mean()
print("\n4. Average salary for each department:")
print(dept_avg_salary)


# ==========================================
# QUESTION 5: SORTING AUR CSV MEIN SAVE KARNA
# ==========================================

# sort_values() se hum salary ko chote se bade (ascending) order mein lagayenge
sorted_df = df.sort_values(by='Salary', ascending=True)
print("\n5. DataFrame sorted by Salary (Low to High):")
print(sorted_df)

# Ab is result ko ek naye CSV file mein save karenge
# index=False ka matlab hai ki hum extra row numbers save nahi karna chahte
sorted_df.to_csv('sorted_employees.csv', index=False)
print("\nResult has been saved to 'sorted_employees.csv' file.")
```
