```python
import random
import numpy as np
import matplotlib.pyplot as plt

# =========================================================
# QUESTION 1: COIN TOSS & TWO DICE SUM
# =========================================================

# --- Part (a): Coin Toss 10,000 times ---
def q1_coin_toss():
    # Step 1: Define total number of trials (how many times we toss)
    trials = 10000
    heads_count = 0
    tails_count = 0

    # Step 2: Use a loop to repeat the toss 10,000 times
    for i in range(trials):
        # Step 3: Pick 'Heads' or 'Tails' randomly
        result = random.choice(['Heads', 'Tails'])
        if result == 'Heads':
            heads_count += 1
        else:
            tails_count += 1

    # Step 4: Calculate probability (Success / Total)
    prob_heads = heads_count / trials
    prob_tails = tails_count / trials

    print(f"1a. Heads Prob: {prob_heads}, Tails Prob: {prob_tails}")

# --- Part (b): Rolling two dice for sum of 7 ---
def q1_dice_sum():
    # Step 1: Define total trials
    trials = 10000
    sum_7_count = 0

    # Step 2: Loop to roll dice many times
    for i in range(trials):
        # Step 3: Roll two dice (random number between 1 and 6)
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        
        # Step 4: Check if their sum is exactly 7
        if (die1 + die2) == 7:
            sum_7_count += 1

    # Step 5: Calculate probability
    prob_7 = sum_7_count / trials
    print(f"1b. Probability of getting sum 7: {prob_7}")

# =========================================================
# QUESTION 2: AT LEAST ONE '6' IN 10 ROLLS
# =========================================================

def q2_at_least_one_six():
    # Step 1: Total trials (testing the whole experiment 10,000 times)
    trials = 10000
    success_count = 0

    # Step 2: Start the loop for trials
    for i in range(trials):
        # Step 3: Roll one die 10 times and store in a list
        ten_rolls = [random.randint(1, 6) for _ in range(10)]
        
        # Step 4: Check if digit '6' is present at least once in those 10 rolls
        if 6 in ten_rolls:
            success_count += 1

    # Step 5: Final probability calculation
    prob = success_count / trials
    print(f"2. Prob of at least one '6' in 10 rolls: {prob}")

# =========================================================
# QUESTION 3: BALLS SIMULATION (CONDITIONAL PROBABILITY)
# =========================================================

def q3_ball_logic():
    # Step 1: Define the bag with balls (5 Red, 7 Green, 8 Blue)
    bag = ['Red']*5 + ['Green']*7 + ['Blue']*8
    history = []
    trials = 10000

    # Step 2: Simulate drawing a ball and putting it back 10,000 times
    for i in range(trials):
        history.append(random.choice(bag))

    # Step 3: Track how many times 'Blue' was drawn
    blue_occured = 0
    red_after_blue = 0

    # Step 4: Loop through history to find (Blue followed by Red)
    for i in range(len(history) - 1):
        if history[i] == 'Blue':
            blue_occured += 1
            # Step 5: Check if the very next ball was Red
            if history[i+1] == 'Red':
                red_after_blue += 1

    # Step 6: Probability of Red GIVEN that previous was Blue
    cond_prob = red_after_blue / blue_occured
    print(f"3. P(Red | Blue): {cond_prob}")

# =========================================================
# QUESTION 4: DISCRETE RANDOM VARIABLES (NUMPY)
# =========================================================

def q4_discrete_var():
    # Step 1: Define the values (X) and their given probabilities (P)
    values = [1, 2, 3]
    probabilities = [0.25, 0.35, 0.40]

    # Step 2: Use numpy to pick 1000 samples based on these probabilities
    sample = np.random.choice(values, size=1000, p=probabilities)

    # Step 3: Use numpy built-in functions for Mean, Variance, and Std Dev
    mean_val = np.mean(sample)
    var_val = np.var(sample)
    std_val = np.std(sample)

    print(f"4. Mean: {mean_val}, Variance: {var_val}, Std Dev: {std_val}")

# =========================================================
# QUESTION 5: CONTINUOUS RANDOM VARIABLE (EXPONENTIAL)
# =========================================================

def q5_exponential_sim():
    # Step 1: Generate 2000 samples from exponential distribution with mean=5
    # Note: In numpy, 'scale' is the mean.
    samples = np.random.exponential(scale=5, size=2000)

    # Step 2: Create a histogram for visualization
    plt.hist(samples, bins=50, density=True, color='lightgreen', label='Sample Data')

    # Step 3: Create the PDF (Probability Density Function) line for overlay
    x = np.linspace(0, 25, 100)
    pdf = (1/5) * np.exp(-x/5) # Mathematical formula for exponential PDF
    plt.plot(x, pdf, color='red', label='PDF Line')
    
    plt.title("Exponential Distribution (Mean 5)")
    plt.legend()
    plt.show()

# =========================================================
# QUESTION 6: CENTRAL LIMIT THEOREM (CLT)
# =========================================================

def q6_clt_simulation():
    # Step 1: Generate 10,000 random numbers (Uniform distribution)
    # This data is "Flat" (all numbers from 0 to 10 are equally likely)
    parent_data = np.random.uniform(0, 10, 10000)

    # Step 2: Create a list to store the 'means' of our samples
    sample_means = []

    # Step 3: Take 1000 different samples, each of size 30
    for i in range(1000):
        # Pick 30 random numbers from our parent data
        small_sample = np.random.choice(parent_data, size=30)
        # Calculate the average (mean) of these 30 numbers
        sample_means.append(np.mean(small_sample))

    # Step 4: Visualize both to compare
    plt.figure(figsize=(10, 4))
    
    # Left plot: The flat original data
    plt.subplot(1, 2, 1)
    plt.hist(parent_data, bins=20, color='gray')
    plt.title("Original Uniform Data")

    # Right plot: The distribution of means (should look like a Bell Curve)
    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=30, color='blue')
    plt.title("Distribution of Sample Means (CLT)")

    plt.show()

# --- Calling all functions to see the output ---
q1_coin_toss()
q1_dice_sum()
q2_at_least_one_six()
q3_ball_logic()
q4_discrete_var()
q5_exponential_sim()
q6_clt_simulation()
```
